```python
import numpy as np
import pandas as pd
import textdistance 
from collections import Counter
import re
```

```python
words = []
with open('autocorrect book.txt','r',encoding='utf-8') as f:
    data = f.read()
    data = data.lower()
    word = re.findall('\w+', data)
    words +=word
```

```python
word_freq_dict = Counter(words)
```

```python
word_freq_dict
```

```python
word_freq_dict.most_common(10)
```

```python
Total_words_freq = sum(word_freq_dict.values())


probs = {}
for k in word_freq_dict.keys():
    probs[k] = word_freq_dict[k] / Total_words_freq
```

```python
Total_words_freq
```

```python
probs
```

```python
def autocorrect(word): 
    word = word.lower() 
    if word in probs:
        print('The word is present in the document :', word)
    else:
        similarities = [1-(textdistance.Jaccard(qval=2)).distance(w,word) for w in word_freq_dict.keys()]
        df = pd.DataFrame.from_dict(probs,orient='index').reset_index()
        df = df.rename(columns={'index':'Word',0:'Prob'})
        df['Similarity'] = similarities
        output = df.sort_values(['Similarity','Prob'],ascending=False).head(10)
        return(output)
```
